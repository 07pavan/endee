dataset loading

```python
import json
import os

file_path = "C:\\Users\\91866\\endee\\Govt_Scheme_Copilot\\data\\schemes.json" 

try:
    with open(file_path, "r") as f:
        data = json.load(f)
    print(len(data))
    print(data[0])
except FileNotFoundError:
    print(f"File not found: {file_path}. Current working directory: {os.getcwd()}")
    print("Check the path or create the file/directory.")
except json.JSONDecodeError as e:
    print(f"JSON decode error: {e}")
```

Load model

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")
```

Generate embeddings

```python
texts = [item["text"] for item in data]
embeddings = model.encode(texts)

print(len(embeddings))
```

Define search function 

```python
import numpy as np

def search(query, top_k=3):
    query_embedding = model.encode([query])[0]
    
    similarities = np.dot(embeddings, query_embedding)
    
    top_indices = np.argsort(similarities)[-top_k:][::-1]
    
    results = [data[i]["name"] for i in top_indices]
    return results
```

Test

```python
query = "I am a farmer with low income"
results = search(query)

print(results)
```
